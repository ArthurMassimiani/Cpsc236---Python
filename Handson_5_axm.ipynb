## 7.2 Wizard Inventory
Create a program that keeps track of the items that a wizard can carry.

### Console:
```powershell
The Wizard Inventroy Program

COMMAND MENU
walk - Walk down the path 
show - Show all items 
drop - Drop an item
exit - Exit program

Command: walk
While walking down a path, you see a scroll of uncursing. 
Do you want to grab it? (y/n): y
You picked up a scroll of uncursing.

Command: walk
While walking down a path, you see an unknown potion. 
Do you want to grab it? (y/n): y
You can't carry any more items. Drop something first.

Command: show
1.	a wooden staff
2.	a scroll of invisibility
3.	a crossbow
4.	a scroll of uncursing

Command: drop Number: 3
You dropped a crossbow.

Command: exit 
Bye!
```

### Specifications:
- Use the attached text file named `wizard_all_items.txt` that contains a list of all the items that a wizard can carry. (in the corresponding directory)
- When the user selects the walk command, the program should read the items from the file, randomly pick one, and give the user the option to grab it.
- Your program should create another file that stores the items that the wizard is carrying (`wizard_inventory.txt`). Make sure to update this file every time the user grabs or drops an item.
- The wizard can only carry **four** items at a time.
- For the drop command, display an error message if the user enters an invalid number for the item.



````python
import random

wizard_inventory_file = "wizard_inventory.txt"
wizard_items_file = "wizard_all_items.txt"

def read_items_from_file(wizard_items_file):
    with open(wizard_items_file, 'r') as file:
        items = file.readlines()
    return [item.strip() for item in items]

def select_random_item(items):
    return random.choice(items)

def drop_item(inventory_file, item_number):
    try:
        with open(inventory_file, 'r') as file:
            items = file.readlines()
        del items[item_number - 1]
        with open(inventory_file, 'w') as file:
            file.writelines(items)
        print("Item dropped successfully.")
    except (IndexError, FileNotFoundError):
        print("Invalid item number.")

def main():
    command_menu = (
        "COMMAND MENU\n"
        "walk - Walk down the path\n"
        "show - Show all items\n"
        "drop - Drop an item\n"
        "exit - Exit program\n"
    )

    print("The Wizard Inventory Program\n")
    print(command_menu)

    while True:
        command = input("\nCommand: ").lower().strip()

        if command == "walk":
            items = read_items_from_file(wizard_items_file)
            if items:
                random_item = select_random_item(items)
                print(f"\nWhile walking down a path, you see {random_item}.")
                grab = input("Do you want to grab it? (y/n): ").lower().strip()
                if grab == 'y':
                    try:
                        with open(wizard_inventory_file, 'r') as file:
                            if len(file.readlines()) >= 4:
                                print("You can't carry any more items. Drop something first.")
                            else:
                                with open(wizard_inventory_file, 'a') as inventory_file:
                                    inventory_file.write(random_item + '\n')
                                print(f"You picked up {random_item}.")
                    except FileNotFoundError:
                        with open(wizard_inventory_file, 'w') as inventory_file:
                            inventory_file.write(random_item + '\n')
                        print(f"You picked up {random_item}.")
                else:
                    print("You chose not to grab the item.")
            else:
                print("No items found.")
        
        elif command == "show":
            print("\nCurrent inventory:")
            try:
                with open(wizard_inventory_file, 'r') as file:
                    items = file.readlines()
                    if items:
                        for index, item in enumerate(items, 1):
                            print(f"{index}. {item}", end='')
                    else:
                        print("Inventory is empty.")
            except FileNotFoundError:
                print("Inventory is empty.")
        
        elif command == "drop":
            try:
                print("\nCurrent inventory:")
                with open(wizard_inventory_file, 'r') as file:
                    items = file.readlines()
                    if items:
                        for index, item in enumerate(items, 1):
                            print(f"{index}. {item}", end='')
                    else:
                        print("Inventory is empty.")
                item_number = input("\nEnter the number of the item you want to drop: ")
                if item_number.isdigit():
                    drop_item(wizard_inventory_file, int(item_number))
                else:
                    print("Invalid input. Please enter a number.")
            except FileNotFoundError:
                print("Inventory is empty.")
        
        elif command == "exit":
            print("Bye!")
            break
        
        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
````

## 7.3 Monthly Sales
Create a program that reads the sales for 12 months from a file and calculates the total yearly sales as well as the average monthly sales. In addition, this program should let the user edit the sales for any month.

### Console:

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesn’t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
import csv
import os

def load_sales_data(file_name):
    sales_data = {}
    with open(file_name, newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            month, sales = row
            sales_data[month] = int(sales)
    return sales_data

def save_sales_data(file_name, sales_data):
    with open(file_name, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        for month, sales in sales_data.items():
            writer.writerow([month, sales])

def calculate_summary(sales_data):
    yearly_total = sum(sales_data.values())
    monthly_average = yearly_total / len(sales_data)
    return yearly_total, monthly_average

def edit_sales(sales_data, month, new_sales):
    if month in sales_data:
        sales_data[month] = new_sales
        return True
    else:
        return False

def main():
    file_name = "monthly_sales.csv"
    sales_data = load_sales_data(file_name)
    
    print("\nCOMMAND MENU")
    print("monthly - View monthly sales")
    print("yearly  - View yearly summary")
    print("edit    - Edit sales for a month")
    print("exit    - Exit program")

    while True:
        command = input("\nCommand: ").strip().lower()

        if command == "monthly":
            print("\nMonthly Sales program:")
            for month, sales in sales_data.items():
                print(f"{month} - {sales}")

        elif command == "yearly":
            yearly_total, monthly_average = calculate_summary(sales_data)
            print(f"\nYearly total:   {yearly_total}")
            print(f"Monthly average:    {monthly_average:.2f}")

        elif command == "edit":
            month = input("Three-letter Month: ").strip().capitalize()[:3]
            new_sales = int(input("Sales Amount: "))
            if edit_sales(sales_data, month, new_sales):
                save_sales_data(file_name, sales_data)
                print(f"Sales amount for {month} was modified.")
            else:
                print("Invalid month abbreviation.")

        elif command == "exit":
            print("Bye!")
            break

        else:
            print("Invalid command. Please try again.")

if __name__ == "__main__":
    main()
````
